### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
#your code here
speed = pokemon[(pokemon['Generation'] == 4) & (pokemon['Legendary'] == False) & (pokemon['Type 1'] == 'Ghost')].sort_values('Speed', ascending = False)
speed.iloc[1]
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
#your code here
charizard_attack = pokemon.loc['Charizard', 'Attack']
charizard_defense = pokemon.loc['Charizard', 'Defense']
pokemon[(pokemon['Legendary'] == False) & (pokemon['Attack'] > charizard_attack) & (pokemon['Defense'] > charizard_defense)].shape[0]
```
